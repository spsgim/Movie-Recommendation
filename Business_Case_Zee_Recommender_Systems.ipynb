# ***Building Recommendation System***

## Data Preprocessing

The first step is to transform our data into a user-item matrix, also known as a "utility" matrix. In this matrix, rows represent users and columns represent movies. The beauty of collaborative filtering is that it doesn't require any information about the users or the movies user to generate recommendations.

## User-Item (utility) Matrix / Pivot Table

### CSR

````python
M = final_ratings['UserID'].nunique()
N = final_ratings['Movie_Title'].nunique()

user_mapper = dict(zip(np.unique(final_ratings['UserID']), list(range(M))))
movie_mapper = dict(zip(np.unique(final_ratings['Movie_Title']), list(range(N))))

user_inv_mapper = dict(zip(list(range(M)), np.unique(final_ratings['UserID'])))
movie_inv_mapper = dict(zip(list(range(N)), np.unique(final_ratings['Movie_Title'])))

user_index = [user_mapper[i] for i in final_ratings['UserID']]
item_index = [movie_mapper[i] for i in final_ratings['Movie_Title']]

# Storing the Matrix in CSR
X = csr_matrix((final_ratings["Rating"], (user_index,item_index)), shape=(M,N))
````

## ***Item-Item Recommendations with Pearson Correlation***

````python
# Calculate the Pearson correlation matrix for items (movies)
item_correlation = pt.T.corr(method='pearson')

def get_similar_movies_pc(movie_name, num_recommendations=5):
    if movie_name not in item_correlation:
        print(f"Movie '{movie_name}' not found in the dataset.")
        return []

    # Get the correlation series for the input movie
    similar_scores = item_correlation[movie_name]

    # Sort the correlation scores in descending order and get the top `num_recommendations` + 1
    # (including the movie itself)
    similar_movies = similar_scores.sort_values(ascending=False).index.tolist()

    # Exclude the input movie from the recommendations
    similar_movies.remove(movie_name)

    return similar_movies[:num_recommendations]
````

````python
# Example usage
movie_title = "Star Wars: Episode IV - A New Hope"
recommended_movies = get_similar_movies_pc(movie_title)
print(f"Since you watched {movie_title} Recommended movies:")
for i, movie in enumerate(recommended_movies, start=1):
    print(f"{i}. {movie}")
````

### Model

````python
def get_similar_movies_knn(Movie_Title, X=X_scaled.T, movie_mapper=movie_mapper, movie_inv_mapper=movie_inv_mapper, n_recommendations=5, metric='cosine'):
    """
    Finds k-nearest neighbours for a given movie id.

    Args:
        movie_id: id of the movie of interest
        X: user-item utility matrix
        k: number of similar movies to retrieve
        metric: distance metric for kNN calculations

    Output: returns list of k similar movie ID's
    """
    neighbour_ids = []

    movie_ind = movie_mapper[Movie_Title]
    movie_vec = X[movie_ind]
    if isinstance(movie_vec, (np.ndarray)):
        movie_vec = movie_vec.reshape(1,-1)
    # use k+1 since kNN output includes the movieId of interest
    kNN = NearestNeighbors(n_neighbors=n_recommendations+1, algorithm="brute", metric=metric)
    kNN.fit(X)
    neighbour = kNN.kneighbors(movie_vec, return_distance=False)
    for i in range(0,n_recommendations+1):
        n = neighbour.item(i)
        neighbour_ids.append(movie_inv_mapper[n])
    neighbour_ids.pop(0)
    return neighbour_ids
````

## Handling Cold Start Problem - Content Based Filtering

Collaborative filtering relies solely on user-item interactions within the utility matrix. The issue with this approach is that brand new users or items with no iteractions get excluded from the recommendation system. This is called the cold start problem. Content-based filtering is a way to handle this problem by generating recommendations based on user and item features.

````python
cosine_sim_genre = cosine_similarity(movie_genres, movie_genres)
print(f"Dimensions of our genres cosine similarity matrix: {cosine_sim_genre.shape}")
````

````python
def movie_finder(title):
    all_titles = movies['Movie_Title'].tolist()
    closest_match = process.extractOne(title,all_titles)
    return closest_match[0]
````

````python
movie_idx = dict(zip(movies['Movie_Title'], list(movies.index)))
````

````python
def get_content_based_recommendations(title_string, n_recommendations=10):

    title = movie_finder(title_string)
    idx = movie_idx[title]

    sim_scores = list(enumerate(cosine_sim_genre[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:(n_recommendations+1)]

    similar_movies = [i[0] for i in sim_scores]

    print(f"Because you watched {title}:")
    print(movies['Movie_Title'].iloc[similar_movies])
````

````python
def get_rec_svd(user_id, n_recommendations = 5):
    """Returns a list of n_recommendations for a given user"""

    # It doesn't make sense to recommend a movies which user has already recommended
    # So, we are subracting those movies from all movies to get the list of movies to predict from

    user_movies = ratings_df[ratings_df['UserID'] == user_id]['Movie_Title'].unique() # user-item interactions
    all_movies = ratings_df['Movie_Title'].unique() # all-items (movies)
    movies_to_predict = list(set(all_movies) - set(user_movies)) # items users doesn't interacted

    user_movie_pairs = [(user_id, movie_id, 0) for movie_id in movies_to_predict] #Creating a tuple of uid, iid, r_ui for predictions
    preds = svd.test(user_movie_pairs) # making predictions

    # Predictions Example:  Prediction(uid=2, iid='Gung Ho', r_ui=0, est=3.1224357589473897, details={'was_impossible': False})
    # Sorting based on est
    top_est = sorted(preds, key = lambda x: x.est)[:n_recommendations]

    recommendations = []
    for pred in top_est:
        recommendations.append(pred.iid)
    return recommendations
````

## Embeddings

````python
# Get the item and user embeddings (latent factors)
item_factors = svd.qi
user_factors = svd.pu

# Create a DataFrame for item factors for better handling
item_factors_df = pd.DataFrame(item_factors, index=svd.trainset.all_items())

# Create a DataFrame for user factors for better handling
user_factors_df = pd.DataFrame(user_factors, index=svd.trainset.all_users())
````

````python
# Function to get similar items based on embeddings
def get_similar_items(item_id, num_recommendations=5):
    item_embedding = item_factors_df.loc[item_id]
    similarities = item_factors_df.dot(item_embedding)
    similar_items = similarities.sort_values(ascending=False).index[1:num_recommendations+1]
    return similar_items

# Function to get similar users based on embeddings
def get_similar_users(user_id, num_recommendations=5):
    user_embedding = user_factors_df.loc[user_id]
    similarities = user_factors_df.dot(user_embedding)
    similar_users = similarities.sort_values(ascending=False).index[1:num_recommendations+1]
    return similar_users
````

## **10.Give the sparse 'row' matrix representation for the following dense matrix -**
```
[[1 0]
[3 7]]
```

- ***Sparse 'row' matrix representation***

```
Data:    [1, 3, 7]
Indices: [0, 0, 1]
Indptr:  [0, 1, 3]
```

The sparse 'row' matrix representation, also known as the Compressed Sparse Row (CSR) format, is a way to represent a sparse matrix efficiently by storing only the non-zero elements along with their row and column indices.

For the dense matrix:
```
[[1 0]
 [3 7]]
```

The CSR representation consists of three arrays:
1. **Data**: Contains the non-zero elements of the matrix.
2. **Indices**: Contains the column indices of the corresponding elements in the data array.
3. **Indptr**: Contains the index pointers to the start of each row in the data and indices arrays.


### Step-by-Step Construction of CSR Representation

1. **Data** Array:
   - List all non-zero elements in row-major order.
   - Data: `[1, 3, 7]`

2. **Indices** Array:
   - List the column indices for each non-zero element.
   - Indices: `[0, 0, 1]`

3. **Indptr** Array:
   - List the indices in the data array that start a new row.
   - Indptr: `[0, 1, 3]`
   - Here, `0` indicates the start of the first row, `1` indicates the start of the second row, and `3` is the end of the data array.

### CSR Representation
```
Data:    [1, 3, 7]
Indices: [0, 0, 1]
Indptr:  [0, 1, 3]
```


````python
from scipy.sparse import csr_matrix

# Define the dense matrix
dense_matrix = np.array([[1, 0],
                         [3, 7]])

# Convert to CSR format
csr = csr_matrix(dense_matrix)

print("Data:", csr.data)
print("Indices:", csr.indices)
print("Indptr:", csr.indptr)
````

So, the sparse 'row' matrix (CSR) representation for the given dense matrix is:
```
Data:    [1, 3, 7]
Indices: [0, 0, 1]
Indptr:  [0, 1, 3]
```
